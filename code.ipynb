```python
# Bar plot to show city wise customer distriubution
plt.figure(figsize = (14,6))
ax = sns.countplot(x = "City", data = data, palette = "tab20")
ax.set_title(label = "City ", fontsize= 25)
ax.set_xlabel(xlabel = "City Name", fontsize = 16)
ax.set_ylabel(ylabel = " Customer Count", fontsize = 16)
# we can see there is almost equal distribution of customers in the city
```

```python
#what is the city wise distribution customer gender
#city_income_df = data.groupby('City')[['City','Unit price','Total','gross income']].sum()
plt.figure(figsize = (14,6))
plt.style.use('classic')
ax = sns.countplot(x="City", hue = "Gender", data = data, palette= "tab20")
ax.set_title(label = "Gender Distribution in Cities", fontsize= 25)
ax.set_xlabel(xlabel = "Gender", fontsize = 16)
ax.set_ylabel(ylabel = "Peple Count", fontsize = 16)
```

```python
#distribution of prefered productline in cities
plt.figure(figsize = (20,15))
plt.style.use('classic')
ax = sns.countplot(x="City", hue = "Product line", data = data, palette= "rocket_r")
ax.set_title(label = "Gender Distribution in Cities", fontsize= 25)
ax.set_xlabel(xlabel = "Gender", fontsize = 16)
ax.set_ylabel(ylabel = "Peple Count", fontsize = 16)
```
